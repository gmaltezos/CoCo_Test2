## Pancreas Dynamics, State, and Action Spaces

A block diagram of the system dynamics is depicted on the next figure:

![Imae](modelPNG_draft.png)

As already explained, the focus of the problem is to use the existent inputs to the system, in order to keep the glucose line inside the predetermined limits.
### Actions
Two inputs are available for regulating the levels of glucose:
- $i(t)$: Exogenous Insulin (U);
- $m(t)$: Carbohydrate Intake (U).

The two inputs are provided to the simulation as an action:

$$\textbf{u} = \begin{bmatrix}i(t) & m(t) \end{bmatrix}$$

Of course, your controller can only change the exogenous insulin. The carbohydrate intake comes from the meals and is part of the data describing the patient. Basically, as you will see later, you are going to be able to control only the basal insulin intake, which is the one that actual pumping mechanisms are controling, when reading the measurements from the CGM sensors, every 10 minutes.

<font color='red'>NOTE</font>: These two variables could also be seen in bibliography as I(t) or basal_insulin for the Exogenous Insulin and cho(t) for the Carbohydrate Intake. For keeping some coding best practices, we named these variables with a simple small letter.

### State Variables
The state of the system is defined by the following variables:
- $I_{sc1}(t)$: Insulin in a non-monomeric state (mU/kg);
- $I_{sc2}(t)$: Insulin in a monomeric state (mU/kg);
- $I_p(t)$: Plasma insulin concentration (mU/I);
- $Q_{sto1}(t)$: Amount of insulin in the stomach in solid state (mg/kg);
- $Q_{sto2}(t)$: Amount of insulin in the stomach in liquid state (mg/kg);
- $Q_{gut}(t)$: Glucose concentration in the intestine (mg/kg);
- $G(t)$: Plasma glucose concentration (mg/dl); 
- $X(t)$: Insulin action on glucose disposal and production ($min^{-1}$);
- $IG(t)$: interstitial glucose concentration (mg/dl);

<font color='red'>NOTE</font>: The ReplayBG simulator uses 3 states to represent the meals ($Q_{sto1}(t)$, $Q_{sto2}(t)$, $Q_{gut}(t)$). That way, if we have 3 meals, we get 9  states that describe the meals. We simplified this by averaging the values for the meals, so that now, we have only 3 states for the meals, as described above.

### System Equations
The Non-Linear Differential Equations which describe the behavior of the system are the following:

$\dot{I}_{sc1}(t) = -k_{d}\cdot I_{sc1}(t) + i(t-\beta / V_{I})$

$\dot{I}_{sc2}(t) = k_{d}\cdot I_{sc1}(t) - k_{\alpha2} \cdot I_{sc2}(t)$

$\dot{I}_{p}(t) = k_{\alpha2} \cdot I_{sc2}(t) - k_{e} \cdot I_{p}(t)$

$\dot{Q}_{sto1}(t) = -k_{empt} \cdot Q_{sto1}(t) + m(t)$

$\dot{Q}_{sto2}(t) = k_{empt} \cdot Q_{sto1}(t) - k_{empt} \cdot Q_{sto2}(t)$

$\dot{Q}_{gut}(t) = k_{empt} \cdot Q_{sto2}(t) - k_{abs} \cdot Q_{gut}(t)$

$\dot{G}(t) = -[SG + \rho(G)X(t)] \cdot G(t) + SG \cdot G_{b} + R_{\alpha}(t)/V_G$

$\dot{X}(t) = -p_2 \cdot [X(t) - SI \cdot (I_p(t) - I_pb)]$

$\dot{IG}(t) = -\frac{1}{\alpha}(IG(t) - G(t))$

where, $k_{d}, k_{\alpha2}, k_e, \beta, k_{empt}, k_{abs}, \alpha, p_2$ are parameters describing the rate of diffusion and absorption of insulin, $V_{I}$ is the volume of insulin distribution, $R_{\alpha}(t)$ is the rate of glucose appearance in plasma, $SG$ is the glucose effectiveness, $G_b$ is the basal glucose concentration in the plasma, $V_G$ is the volume of glucose distribution, $I_{pb}$ is the basal insulin concentration in the plasma, $SI$ is the insulin sensitivity, $\rho(G)$ is a function to better represent glucose dynamics in the hypoglycemic range. The first 3 equations describe the absorption dynamics of exogenous insulin infusion to the plasma, the next 3 equations describe the oral glucose absorption system, so they refer to the meals and the last 3 equations describe the glucose-insulin kinetics subsystem. For more detailed information on this system, you can read the corresponding chapter on the [ReplayBg paper](https://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=10164140).
  
### Linearized Equations
A basic linearized model of the dynamics has been derived and is provided to you, for the discrete-time update equation of the form:

<!-- $$\textbf{x}_{k+1} = A \textbf{x}_{k} + B\textbf{i}_{k} + M\textbf{m}_{k} + C.$$ -->
$$\textbf{x}_{k+1} = A \textbf{x}_{k} + B\textbf{u}_{k} + C.$$
<!-- $$\dot{\textbf{x}} = A \textbf{x} + B\textbf{u}$$ -->
 <!-- $$\textbf{x}_{t+1} = A \textbf{x}_{t} + B\textbf{u}_{t}.$$ -->
 
Note:
- The $A$ matrix has shape $\mathbb{R}^{9\times9}$: State matrix;
- The $B$ matrix has shape $\mathbb{R}^{9\times2}$: It is used to include the inputs. Basically, this matrix is separated into two ${9\times1}$ matrices, matrix $I$, which handles the actual basal_insulin input and $M$ which handles the meals;
<!-- - The $M$ matrix has shape $\mathbb{R}^{9\times1}$: It is used to include the meal input; -->
- The $C$ matrix has shape $\mathbb{R}^{9\times1}$: It contains a constant which comes form the linearization;
- We linearize around a single state $xss = [120.0, 0.0016, 0.0, 0.0, 0.0, 1.5081, 10.2735, 0.2827, 120.0]$.

<font color='red'>NOTE</font>: As already said, although the output has two variables, we are able to control only one of them, namely the basal rate. The second variable is the meal input, which we cannot control and depends on the patient.

The function that handles the linearization can be seen on the next cells for better understanding the system.

````python
# Default parameters of the system. These are used throughout the simulation

class ModelParameters:

    def __init__(self, data, bw, is_single_meal):

        """
        Function that returns the default parameters values of the model.

        Parameters
        ----------
        data : pd.DataFrame
                Pandas dataframe which contains the data to be used by the tool.
        bw : double
            The patient's body weight.

        Returns
        -------
        model_parameters: dict
            A dictionary containing the default model parameters.

        Raises
        ------
        None

        See Also
        --------
        None

        Examples
        --------
        None
        """
        # Initial conditions
        self.Xpb = 0.0  # Insulin action initial condition
        self.Qgutb = 0.0  # Intestinal content initial condition

        # Glucose-insulin submodel parameters
        self.VG = 1.45  # dl/kg
        self.SG = 2.5e-2  # 1/min
        self.Gb = 119.13  # mg/dL
        self.r1 = 1.4407  # unitless
        self.r2 = 0.8124  # unitless
        self.alpha = 7  # 1/min
        # All meals use the same SI
        self.SI = 10.35e-4 / self.VG  # mL/(uU*min)
        self.SI_B = 10.35e-4 / self.VG  # mL/(uU*min)
        self.SI_L = 10.35e-4 / self.VG  # mL/(uU*min)
        self.SI_D = 10.35e-4 / self.VG  # mL/(uU*min)
        self.p2 = 0.012  # 1/min
        self.u2ss = np.mean(data.basal) * 1000 / bw  # mU/(kg*min)

        # Subcutaneous insulin absorption submodel parameters
        self.VI = 0.126  # L/kg
        self.ke = 0.127  # 1/min
        self.kd = 0.026  # 1/min
        # model_parameters['ka1'] = 0.0034  # 1/min (virtually 0 in 77% of the cases)
        #self.ka1 = 0.0
        self.ka2 = 0.014  # 1/min
        self.tau = 8  # min
        self.Ipb = self.ke * self.u2ss / self.kd + (self.ka2 / self.ke) * (self.kd / self.ka2) * self.u2ss / self.kd  # from eq. 5 steady-state

        # Oral glucose absorption submodel parameters
        self.kabs = 0.012  # 1/min
        self.beta = 0  # 1/min
        self.kabs_B = 0.012  # 1/min
        self.kabs_L = 0.012  # 1/min
        self.kabs_D = 0.012  # 1/min
        self.kabs_S = 0.012  # 1/min
        self.kabs_H = 0.012  # 1/min
        self.beta_B = 0  # min
        self.beta_L = 0  # min
        self.beta_D = 0  # min
        self.beta_S = 0  # min
        self.kgri = 0.18  # = kmax % 1/min
        self.kempt = 0.18  # 1/min
        self.f = 0.9  # dimensionless
        self.kabs_avg = (self.kabs_B+self.kabs_L+self.kabs_D)/3

        # Exercise submodel parameters
        self.VO2rest = 0.33  # dimensionless, VO2rest was derived from heart rate round(66/(220-30))
        self.VO2max = 1  # dimensionless, VO2max is normalized and estimated from heart rate (220-age) = 100%.
        self.e1 = 1.6  # dimensionless
        self.e2 = 0.78  # dimensionless

        # Patient specific parameters
        self.bw = bw  # kg
        self.to_g = self.bw / 1000
        self.to_mgkg = 1000 / self.bw

        # Measurement noise specifics
        self.SDn = 5

        # Initial conditions
        if 'glucose' in data:
            idx = np.where(data.glucose.isnull().values == False)[0][0]
            self.G0 = data.glucose[idx]
        else:
            self.G0 = self.Gb
````

````python
# Function for linearization
def get_state_equations_example(T, x):
        """
        Parameters
        ----------
        T : discretization step
        Returns
        -------
        A, B, C : discretized linear state equations
        """
         # Matrix Initialization
        Alinear = np.empty([9, 9])
        Blinear = np.empty([9, ])
        Clinear = np.empty([9, ])
        Mlinear = np.empty([9, ])
    
        x0 = x[0]
        x1 = x[1]

        # mp = model_parameters

        k1_T = 1 / (1 + T*mp.kgri)
        k2_T = T*mp.kgri / (1 + T*mp.kempt)
        k3_T = 1 / (1 + T*mp.kempt)
        kb_T = 1 / (1 + T*mp.kabs_avg)
        ki1_T = 1 / (1 + T*mp.kd)
        ki2_T = 1 / (1 + T*mp.ka2)
        kie_T = 1 / (1 + T*mp.ke)
        # print("kabs: ", mp.kabs)
        # print("kabs_B: ", mp.kabs_B)
        # print("kabs_L: ", mp.kabs_L)
        # print("kabs_D: ", mp.kabs_D)
        # print("kabs_S: ", mp.kabs_S)
        # print("kabs_H: ", mp.kabs_H)
        # assert(0)
        x1_c_T = mp.p2 * mp.SI / mp.VI * T/(1+mp.p2*T)
        x4_c_T = T*mp.kempt * kb_T
        x6_c_T = T*mp.kd * ki2_T
        x7_c_T = T*mp.ka2 * kie_T

        df_dx0 = -mp.SG - x1 * (1 + mp.r1)
        df_dx1 = -x0 * (1 + mp.r1)
        df_dx4 = (mp.f * mp.kabs_avg) / mp.VG
        dx0_dt = -mp.SG*x0 -x0*x1*(1 + mp.r1) + mp.SG*mp.Gb
        div = (1-T*df_dx0)

        #                      x0               x1            x2      x3         x4        x5           x6            x7         x8
        Alinear[:] =  [ [1/div         , T*df_dx1/div , 0   , 0     , T*df_dx4/div, 0     , 0             , 0           , 0             ],
                             [0             , 1/(1+mp.p2*T), 0   , 0     , 0           , 0     , x1_c_T*x7_c_T , x1_c_T*kie_T, 0             ],
                             [0             , 0            , k1_T, 0     , 0           , 0     , 0             , 0           , 0             ],
                             [0             , 0            , k2_T, k3_T  , 0           , 0     , 0             , 0           , 0             ],
                             [0             , 0            , 0   , x4_c_T, kb_T        , 0     , 0             , 0           , 0             ],
                             [0             , 0            , 0   , 0     , 0           , ki1_T , 0             , 0           , 0             ],
                             [0             , 0            , 0   , 0     , 0           , x6_c_T, ki2_T         , 0           , 0             ],
                             [0             , 0            , 0   , 0     , 0           , 0     , x7_c_T        , kie_T       , 0             ], 
                             [T/(T+mp.alpha), 0            , 0   , 0     , 0           , 0     , 0             , 0           , mp.alpha/(T+mp.alpha)]]

        # TODO: add meal effect
        Blinear[:] = [0, 0, 0, 0, 0, T/(1+T*mp.kd)*mp.to_mgkg, 0, 0, 0]
        Mlinear[:] = [0, 0, k1_T, 0, 0, 0, 0, 0, 0]
        Clinear[:] = [ T*(dx0_dt - x0*df_dx0 - x1*df_dx1)/div, 0, 0, 0, 0, 0, 0, 0, 0]

        return Alinear, np.vstack((Blinear, Mlinear)).T, Clinear
````

## PID Controller

The following code initializes a basic PID-type controller that can be used to regulate glucose for very specific patients and initial conditions. 

````python
# Initialization
integral = 0
prev_error = 0

def PID(model, x0, tau, time_index, inputs_, xss, uss, dss):
    global integral, prev_error
    
    # Error calculation from the desired setpoint
    error = x0[8]-xss[8]
    
    # Integral, Derivative and Proportional terms of PID
    integral += error*1
    derivative = (error-prev_error)*1
    prev_error = error
    
    # PID gains
    Ki = 0.0
    Kd = 0.005
    Kp = 0.0002


    return max(0, Ki*integral + Kd*derivative + Kp*error + uss), dss 
integral = 0
prev_error = 0
````
