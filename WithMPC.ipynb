## Introduction

Humans produce glucose from different food sources for their cells to use as a source of energy. Glucose travels in the bloodstream and it is captured by insulin, a hormone that moves glucose from the blood into the cells to achieve this transfer of energy. It is imperative, that glucose stays inside specific limits. Too high or too low levels of glucose in the blood could lead to severe implications to a person's health. People with diabetes have higher-than-normal levels of glucose in their blood. Either they don't have enough insulin to move it through or their cells don't respond to insulin as well as they should.

The goal of this project is to work with this simulator and to develop controllers which are able to achieve better results than the already implemented PID controller, so that the glucose levels are regulated in the limits. The behavior of your controllers can be seen through the produced figures and through different metrics (Time-in-range, Time-above-range, Time-below-range).

## Pancreas Dynamics, State, and Action Spaces

A typical Glucose - Time Diagram is depicted on the next figure:



As already explained, the focus of the problem is to use the existent inputs to the system, in order to keep the glucose line inside the predetermined limits.
### Actions
Two inputs are available for regulating the levels of glucose:
- $I(t)$: Exogenous Insulin (U);
- $CHO(t)$: Carbohydrate Intake (U).

The two inputs are provided to the environment as an action:

$$\textbf{u} = \begin{bmatrix}I(t) & CHO(t) \end{bmatrix}$$

Of course, your controller can only change the exogenous insulin. The carbohydrate intake comes from the meals and is part of the data describing the patient.

### State Variables
The state of the system is defined by the following variables:
- $I_{sc1}(t)$: Insulin in a non-monomeric state (mU/kg);
- $I_{sc2}(t)$: Insulin in a monomeric state (mU/kg);
- $I_p(t)$: Plasma insulin concentration (mU/I);
- $Q_{sto1}(t)$: Amount of insulin in the stomach in solid state (mg/kg);
- $Q_{sto2}(t)$: Amount of insulin in the stomach in liquid state (mg/kg);
- $Q_{gut}(t)$: Glucose concentration in the intestine (mg/kg);
- $G(t)$: Plasma glucose concentration (mg/dl); 
- $X(t)$: Insulin action on glucose disposal and production ($min^{-1}$);
- $IG(t)$: interstitial glucose concentration (mg/dl);
  
### Linearized Equations
A basic linearized model of the dynamics has been derived and is provided to you, for the discrete-time update equation of the form:

$$\textbf{x}_{k+1} = A \textbf{x}_{k} + B\textbf{u}_{k}.$$
<!-- $$\dot{\textbf{x}} = A \textbf{x} + B\textbf{u}$$ -->
 <!-- $$\textbf{x}_{t+1} = A \textbf{x}_{t} + B\textbf{u}_{t}.$$ -->
 
Note:
- The $A$ matrix has shape $\mathbb{R}^{9\times 9}$;
- The $B$ matrix has shape $\mathbb{R}^{9\times1}$;
- We linearize around the upright equilibrium with action $\tilde{\textbf{u}} = \begin{bmatrix} 0 & 0 \end{bmatrix}$

## Simulator Usage
To better understand the problem, we are providing a really simple scenario with the most basic basal insulin handler. In this case, if Glucose is between the allowed limitis, we do not deliver any basal insulin, otherwise, we deliver a default rate.

````python
def basal_insulin_handler_example(glucose, meal_announcement, meal_type, hypotreatments, bolus, basal, time, time_index, dss):
    """
    Implements the default basal rate controller: "if G < 90 mg/dl, deliver basal = 0; otherwise, deliver basal = default_rate U/min".

    Parameters
    ----------
    glucose: np.ndarray
        An array vector as long the simulation length containing all the simulated glucose concentrations (mg/dl)
        up to time_index. The values after time_index should be ignored.
    meal_announcement: np.ndarray
        An array vector as long the simulation length containing all the meal announcements (g) up to time_index.
        The values after time_index should be ignored.
    meal_type: np.ndarray
        An array of strings as long the simulation length containing the type of each meal.
        If scenario is `single-meal`, labels can be:
            - `M`: main meal
            - `O`: other meal
        If scenario is `multi-meal`, labels can be:
            - `B`: breakfast
            - `L`: lunch
            - `D`: dinner
            - `S`: snack
            - `H`: hypotreatment
        The values after time_index should be ignored.
    hypotreatments: np.ndarray
        An array vector as long the simulation length containing all the hypotreatment intakes (g/min) up to time_index.
        If the scenario is single meal, hypotreatments will contain only the hypotreatments generated by this function
        during the simulation. If the scenario is multi-meal, hypotreatments will ALSO contain the hypotreatments already
        present in the given data that labeled as such. The values after time_index should be ignored.
    bolus: np.ndarray
        An array vector as long the simulation length containing all the insulin boluses (U/min) up to time_index.
        The values after time_index should be ignored.
    basal: np.ndarray
        An array vector as long the simulation length containing all the insulin basal (U/min) up to time_index.
        The values after time_index should be ignored.
    time: np.ndarray
        An array vector as long the simulation length containing the time corresponding to the current step (hours) up to time_index.
        The values after time_index should be ignored.
    time_index: int
        The index corresponding to the previous simulation step of the replay simulation.
    dss: DSS
        An object that represents the hyperparameters of the integrated decision support system.

    Returns
    -------
    b: float
        The basal insulin rate to administer at time[time_index+1].
    dss: DSS
        An object that represents the hyperparameters of the integrated decision support system.
        dss is also an output since it contains basal_handler_params that beside being a
        dict that contains the parameters to pass to  this function, it also serves as memory area.
        It is possible to store values inside it and the default_basal_handler function will be able
        to access to them in the next call of the function.

    """

    b = dss.basal_handler_params['default_rate']

    # If G < 90...
    if glucose[time_index] < 90:
        # ...set basal rate to 0
        b = 0

    return b, dss
````

````python
from py_agata.py_agata import Agata
from py_agata.time_in_ranges import time_in_target
from py_agata.utils import glucose_time_vectors_to_dataframe

from plot_results import plot_results

from replay_bg_wrapper import simulate

# You can create a dict() that will be accessible during simulation by your handler
basal_handler_params = dict()
basal_handler_params['default_rate'] = 0.01

glucose, insulin_basal, insulin_bolus, cho, time = simulate(basal_handler=basal_insulin_handler_example,
       basal_handler_params=basal_handler_params,
       meal_input_modulation_factor=1)

# After simulation, you can evaluate glucose trace with AGATA.
# First you need to generate a dataframe that is compatible with AGATA
data = glucose_time_vectors_to_dataframe(glucose=glucose, t=time)

# Thenm glucose trace can be analyzed with Agata().analyze_glucose_profile() if you want to obtain an exhaustive set of metrics...
agata = Agata()
results = agata.analyze_glucose_profile(data)
print(results)

# ... or, if you need just 1 metric, using single targeted functions (e.g., time spent in teh target range)
tit = time_in_target(data)
print('TIT = ' + str(tit) + '%')

#Finally, I prepared a function to plot the simulation traces
plot_results(glucose=glucose,insulin_bolus=insulin_bolus, insulin_basal=insulin_basal, cho=cho, time=time)
````
